```python
import json
import ast
%load_ext autoreload
%autoreload 2
import funciones_etl
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

Extraemos el archiovo en formato json

```python
# ruta
ruta = "../Datasets/output_steam_games.json"
```

Abrimos el archivo

```python
# Se crea una lista vacia, se abre el archivo y se recorre linea por linea para luego aggregarlas a la lista 
fila = []
with open(ruta, encoding= "utf-8") as f:
    for line in f.readlines():
        data = json.loads(line)
        fila.append(data)
```

```python
# Con la lista ya cargada se procede a crear el dataframe
df_games = pd.DataFrame(fila)
df_games
```

Podemos observar que el dataframe tiene 13 columnas y 120445 registros, pero existen filas en los que sus registros entereos son de tipo NaN

```python
# Eliminamos los registros vacios
df_games = df_games.dropna(how='all').reset_index(drop=True)
```

```python
df_games
```

Ya no existen todas las filas entreas con valores nulos, verificamos datos y valores nulos en diferentes columnas

```python
funciones_etl.verificar_datos(df_games)
```

podemos observar que la mayoria de las columnas poseen datos nulos pero ue ademas tenemos una columna "id" que funciona como identificador unico por cada registro, asi que podemos verificar valores duplicados en el dataframe

```python
# Valores duplicados
duplicados = funciones_etl.valores_duplicados(df_games, "id")
duplicados
```

Verificando por si id encontramos cuatro registros duplicados, los cuales procederemos a eliominar apoyandonos en el indice

```python
# Eliminamos los registros por el indice
indices = [74, 13894]
df_games = df_games.drop(indices)
# Verificamos de nuevo los datos duplicados
funciones_etl.valores_duplicados(df_games, ["id"])
```

Eliminamos columnas que a priori son innecesarias para la realizacion del trabajo

```python
columnas = ["app_name", "url", "specs", "tags"]
df_games.drop(columns=columnas, axis=1, inplace=True)
df_games.columns
```

## Columna developer

Como sabemos esta columna tiene datos nulos, asi que procedemos verlos

```python
developer_nulos = funciones_etl.valores_nulos_columna(df_games, "developer")
developer_nulos
```

## Colummna title

esta columna tambien tiene valores nulos, que podremos ver

## Columna genres

Eliminamos los valores nulos de la columna genres, usando explode ya que esta esta formada por una lista

```python
df_games = df_games.explode('genres')
df_games = df_games.dropna(subset=['genres'])
```

## Columna "release_date"

Esta columna no sera necesaria para las funciones, pero antes de eliminarla extraemos el año de lanzamiento

```python
df_games["year"] = df_games["release_date"].apply(funciones_etl.funcion_anio)
```

```python
# Eliminamos la columna
df_games = df_games.drop("release_date", axis=1)
```
