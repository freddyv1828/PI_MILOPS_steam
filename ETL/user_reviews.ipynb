```python
import pandas as pd
import json
import ast
%load_ext autoreload
%autoreload 2
import funciones_etl

import warnings
warnings.filterwarnings("ignore")
```

Abrimos el archivo australian_user_reviews

```python
# Ruta del archivo
ruta = "../Datasets/australian_user_reviews.json"
```

```python
# Abrimos el archivo
fila = []
with open(ruta, encoding="utf-8") as f:
    for line in f.readlines():
        fila.append(ast.literal_eval(line))
```

```python
# Creamos el dataframe
df_reviews = pd.DataFrame(fila)
df_reviews
```

Verificar tipo de datos y valores nulos

```python
funciones_etl.verificar_datos(df_reviews)
```

```python
dupli = funciones_etl.valores_duplicados(df_reviews, "user_id")
dupli
```

```python
dupli.sort_values("user_id")
```

Pasamos a eliminar los registros duplicados conservando la primera ocurrencia

```python
df_reviews = df_reviews.drop_duplicates(subset="user_id", keep="first")
```

```python
# Verificamos de nuevo los duplicados
funciones_etl.valores_duplicados(df_reviews, ["user_id"])
```

Observamos que ya no existen valores duplicados, y pasamos a transformar la columna reviews

```python
# Creamos un dataframe a partir de la columna
df_reviews2 = pd.json_normalize(df_reviews["reviews"])
df_reviews2
```

```python
df_reviews2 = pd.concat([df_reviews[['user_id', 'user_url']], df_reviews2], axis=1)
```

```python
# Intentamos darle sentido transformando columnas en filas
df_reviews2 = pd.melt(df_reviews2, id_vars=['user_id', 'user_url'], 
                       value_vars=list(range(9)),
                       value_name='reviews')
```
