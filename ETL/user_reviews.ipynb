```python
import pandas as pd
import json
import ast
%load_ext autoreload
%autoreload 2
import funciones_etl

import warnings
warnings.filterwarnings("ignore")
```

Abrimos el archivo australian_user_reviews

```python
# Ruta del archivo
ruta = "../Datasets/australian_user_reviews.json"
```

```python
# Abrimos el archivo
fila = []
with open(ruta, encoding="utf-8") as f:
    for line in f.readlines():
        fila.append(ast.literal_eval(line))
```

```python
# Creamos el dataframe
df_reviews = pd.DataFrame(fila)
df_reviews
```

Verificar tipo de datos y valores nulos

```python
funciones_etl.verificar_datos(df_reviews)
```

Nos encontramos con un dataset sin valores nulos hasta ahora

procedemos a verificar valores duplicados

```python
dupli = funciones_etl.valores_duplicados(df_reviews, "user_id")
dupli
```

```python
dupli.sort_values("user_id")
```

Pasamos a eliminar los registros duplicados conservando la primera ocurrencia

```python
df_reviews = df_reviews.drop_duplicates(subset="user_id", keep="first")
```

```python
# Verificamos de nuevo los duplicados
funciones_etl.valores_duplicados(df_reviews, ["user_id"])
```

Observamos que ya no existen valores duplicados, y pasamos a transformar la columna reviews

```python
# Creamos un dataframe a partir de la columna
df_reviews2 = pd.json_normalize(df_reviews["reviews"])
df_reviews2
```

Concatenamos los dataframes

```python
df_reviews2 = pd.concat([df_reviews[['user_id', 'user_url']], df_reviews2], axis=1)
```

Vemos el dataset

```python
df_reviews2
```

Transformamos columnas en filas

```python
# Intentamos darle sentido transformando columnas en filas
df_reviews2 = pd.melt(df_reviews2, id_vars=['user_id', 'user_url'], 
                       value_vars=list(range(9)),
                       value_name='reviews')
```

Observamos el dataframe

```python
df_reviews2
```

podemos observar valores nulos en el dataset

```python
# Eliminamos los valores nulos
df_reviews2 = df_reviews2.dropna()
```

Convertimos cada clave del diccionario en una columna

```python
df_reviews = df_reviews2['reviews'].apply(pd.Series, dtype='object')
df_reviews = df_reviews.add_prefix('reviews_')
```

```python
# Verificamos
df_reviews
```

Necesitamos recuperar las columnas user_id y user_url

```python
# Concatenamos los dataframes
df_reviews = pd.concat([df_reviews2[['user_id', 'user_url']], df_reviews], axis=1)
```

```python
df_reviews
```

Podemos obsevar valores vacios en algunos registros, asi que los vamos a reemplazar por None, para poder manejarlos

```python
# Reemplazar valores vacios por nuloa
df_reviews.replace("", None, inplace=True)
```

Verificamos los valores nulos

```python
funciones_etl.verificar_datos(df_reviews)
```

Podemos ver la cantidad de datos nulos, en cada columna, por ejemplo columna "reviews_funny" es demasiado alta asi que procedmos a eliminar columnas que no utilizaremos, ya sea porque no necesitamos o porque al tener un porcentaje de valores nulos tan alto no nos serviria
